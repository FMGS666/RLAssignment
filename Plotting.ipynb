````python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import re
````

### Expected Sarsa History

````python
def open_json(file_path):
    with open(file_path, "r") as fp:
        return json.load(fp)
````

````python
EX_SARSA_HISTORY_PATH = r"C:\Users\lollo\iCloudDrive\Desktop\DSBA\M2_T2\reinforcement\FlappyBird\TDHistory"

EX_SARSA_HISTORY_FILES = os.listdir(EX_SARSA_HISTORY_PATH)

EX_SARSA_HISTORY_FILE_PATHS = [
    os.path.join(EX_SARSA_HISTORY_PATH, EX_SARSA_HISTORY_FILE) 
        for EX_SARSA_HISTORY_FILE in EX_SARSA_HISTORY_FILES 
]

EXPECTED_SARSA_REWARDS = [
    open_json(EX_SARSA_HISTORY_FILE_PATH)["episodes_lengths"] for 
        EX_SARSA_HISTORY_FILE_PATH in EX_SARSA_HISTORY_FILE_PATHS
]

EXPECTED_SARSA_SCORES = [
    open_json(EX_SARSA_HISTORY_FILE_PATH)["episodes_scores"] for 
        EX_SARSA_HISTORY_FILE_PATH in EX_SARSA_HISTORY_FILE_PATHS
]


EX_SARSA_SCORES = [
    int(re.findall(".*_maxScore(\d*).*", EX_SARSA_HISTORY_FILE)[0]) for 
        EX_SARSA_HISTORY_FILE in EX_SARSA_HISTORY_FILES
]

EX_SARSA_AVG_REWARDS = [
    np.mean(EXPECTED_SARSA_REWARD) for 
        EXPECTED_SARSA_REWARD in EXPECTED_SARSA_REWARDS    
]

EX_SARSA_AVG_SCORES = [
    np.mean(EXPECTED_SARSA_SCORE) for 
        EXPECTED_SARSA_SCORE in EXPECTED_SARSA_SCORES 
]

EX_SARSA_EPSILONS = [
    float(re.findall(".*_eps(0\.\d*).*", EX_SARSA_HISTORY_FILE)[0]) for 
        EX_SARSA_HISTORY_FILE in EX_SARSA_HISTORY_FILES
]

EX_SARSA_GAMMAS = [
    float(re.findall(".*_gamma(1|0\.\d*).*", EX_SARSA_HISTORY_FILE)[0]) for 
        EX_SARSA_HISTORY_FILE in EX_SARSA_HISTORY_FILES
]

EX_SARSA_ALPHAS = [
    float(re.findall(".*_alpha(1|0\.\d*).*", EX_SARSA_HISTORY_FILE)[0]) for 
        EX_SARSA_HISTORY_FILE in EX_SARSA_HISTORY_FILES
]
````

````python
EX_SARSA_EPSILONS_e1 = set([
    idx for idx, val in enumerate(EX_SARSA_EPSILONS) 
        if val == 0.1
])
EX_SARSA_EPSILONS_e2 = set([
    idx for idx, val in enumerate(EX_SARSA_EPSILONS) 
        if val == 0.01
])
````

````python
EX_SARSA_GAMMAS_75 = set([
    idx for idx, val in enumerate(EX_SARSA_GAMMAS) 
        if val == 0.75
])

EX_SARSA_GAMMAS_50 = set([
    idx for idx, val in enumerate(EX_SARSA_GAMMAS) 
        if val == 0.5
])

EX_SARSA_GAMMAS_100 = set([
    idx for idx, val in enumerate(EX_SARSA_GAMMAS) 
        if val == 1.0
])
````

````python
EX_SARSA_ALPHAS_1e3 = set([
    idx for idx, val in enumerate(EX_SARSA_ALPHAS) 
        if val == 0.001
])

EX_SARSA_ALPHAS_50 = set([
    idx for idx, val in enumerate(EX_SARSA_ALPHAS) 
        if val == 0.5
])

EX_SARSA_ALPHAS_1e2 = set([
    idx for idx, val in enumerate(EX_SARSA_ALPHAS) 
        if val == 0.01
])

EX_SARSA_ALPHAS_005 = set([
    idx for idx, val in enumerate(EX_SARSA_ALPHAS) 
        if val == 0.005
])


EX_SARSA_ALPHAS_1 = set([
    idx for idx, val in enumerate(EX_SARSA_ALPHAS) 
        if val == 0.005
])
````

#### EPSILON: 0.1, GAMMA: 0.75

````python
EX_SARSA_EPSILONS_e1_GAMMAS_75 = EX_SARSA_EPSILONS_e2.intersection(EX_SARSA_GAMMAS_100)

ALPHAS_EX_SARSA_EPSILONS_e1_GAMMAS_75 = [
    EX_SARSA_ALPHAS[IDX] for IDX in EX_SARSA_EPSILONS_e1_GAMMAS_75
]

MAXSCORES_EX_SARSA_EPSILONS_e1_GAMMAS_75 = [
    EX_SARSA_SCORES[IDX] for IDX in EX_SARSA_EPSILONS_e1_GAMMAS_75
]

AVG_SCORES_EX_SARSA_EPSILONS_e1_GAMMAS_75 = [
    EX_SARSA_AVG_SCORES[IDX] for IDX in EX_SARSA_EPSILONS_e1_GAMMAS_75
]

AVG_REWARDS_EX_SARSA_EPSILONS_e1_GAMMAS_75 = [
    EX_SARSA_AVG_REWARDS[IDX] for IDX in EX_SARSA_EPSILONS_e1_GAMMAS_75
]
````

````python
ALPHAS_EPSILONS_e1_GAMMAS_75_SORTED, MAXSCORES_EPSILONS_e1_GAMMAS_75_SORTED = zip(*sorted(zip(ALPHAS_EX_SARSA_EPSILONS_e1_GAMMAS_75, MAXSCORES_EX_SARSA_EPSILONS_e1_GAMMAS_75)))
ALPHAS_EPSILONS_e1_GAMMAS_75_SORTED, AVGSCORES_EPSILONS_e1_GAMMAS_75_SORTED = zip(*sorted(zip(ALPHAS_EX_SARSA_EPSILONS_e1_GAMMAS_75, AVG_SCORES_EX_SARSA_EPSILONS_e1_GAMMAS_75)))
ALPHAS_EPSILONS_e1_GAMMAS_75_SORTED, AVGREWARDS_EPSILONS_e1_GAMMAS_75_SORTED = zip(*sorted(zip(ALPHAS_EX_SARSA_EPSILONS_e1_GAMMAS_75, AVG_REWARDS_EX_SARSA_EPSILONS_e1_GAMMAS_75)))
````

````python
plt.plot(ALPHAS_EPSILONS_e1_GAMMAS_75_SORTED, MAXSCORES_EPSILONS_e1_GAMMAS_75_SORTED)

plt.xlabel('Alpha')
plt.ylabel('MaxScore')
plt.title('TD Agent (epsilon: 0.01, gamma: 1.0)')

# show the plot
plt.show()
````

````python
plt.plot(ALPHAS_EPSILONS_e1_GAMMAS_75_SORTED, AVGSCORES_EPSILONS_e1_GAMMAS_75_SORTED)

plt.xlabel('Alpha')
plt.ylabel('AvgScore')
plt.title('TD Agent (epsilon: 0.01, gamma: 1.0)')

# show the plot
plt.show()
````
